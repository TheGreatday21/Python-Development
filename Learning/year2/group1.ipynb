#### OUTLIERS IN DATASCIENCE
- An outlier is a data point that is abnormally higher or lower than the rest of the data in a given data set 

##### Importance of outliers 
- They help us spot errors in data entry and measurement.
- Outliers can help us determine unexpected phenomenon in our data 
- Ignoring outliers completely from our data can sometimes lead to skewed results in our data analysis

##### Sources of outliers 
1. Poor data entry into a dataset.
2. Faulty measurements from equipment.
3. Genuine but rare observations like a 2 headed human being.

- The mishandling of outliers can lead to:<br>
    a, Skewed results<br>
    b, Biased conclusions<br> 
    c, Reduced model reliability<br>

##### Types of outliers 
1. <strong>Global outliers(Point anomalies)</strong>
<p>
    These are outliers where one data point is extremly skewed compared to the rest of the data points in the data set forexample in sample data of average wage for people in a company being 500,000 UGX but one individual is earning 130,000,000 UGX</p>
2. <strong>Contexual outliers (Conditional anomalies)</strong>
<p>
    As the name states thihs are outliers that appear when usually in one condition but are abnormal in another forexample a temperature of 25 degrees celcius is normal for data colleected in summer but abnormal if it is seen in winter time  
    </p>
3. <strong>Collective outliers (Group anomalies)</strong>
<p>
    These are outliers where a group of datapoints all behave unusually even if the individual datapoints seem normal forexample if you are monitoring network traffic and a sudden burst occurs on the network, it may indicate a cyber attack
</p>

##### Impact of Outliers 
<p>These imapct both statistical measurements and machine learning models output</p> 
<p>Statistically  outliers :</p><br>
- Lead to drastic shifts in mean <br>
- Inflate spread making variability seem larger <br>
<p>In machine learning models :</p><br>
- In linear regression skewed coefficients pull the line of best fit towards the outliers <br>
- In clustering centroids are distorted by extreme points <br>
- In Classification , decision boundaries shift unfavourably <br>
- In Neural Networks they can lead to training instability if not properly scaled or handled <br>

#### Strategies for handling outliers 
- Identify the cause of the outliers ie if it human error or genuine unique points observed ina given populationn. <br>
<strong>Options for handling</strong> 
<br>
- Deleting the outliers from the dataset ie irrelevant points <br>
- Transforming the outliers using logs , square root or box-cox transformation to reduce skew <br>
- Imputation - This is by far the commonest method of handling outliers where you replace them with mean or median <br>
- Capping is the process of replacingg extreme values with thresholds .<br>

- The main factors that influence decision making process include  
- Nature of the dataset 
- Goal of analysis (prediction or description )
- Sample size (if you remove data from a small dataset might lead to skewed results)
- Domain knowledge - Does the outlier represent some very important information in the data 

#### UNIVARIATE DATA RELATIONSHIPS
- Univariate data is data containing only one variable or characteristic at a time forexample salaries of workers , heights of students, ages of people in a community. 
- We analyse such data to understand independent variables using measure of central tendency like mean, median , mode and dispersion like range or standard deviation .
- In this type of data theree is no comparison between other variables .

##### Types of Univariate data 
<br>
1. Categorical/Qualitative<br>
    a. Nominal - Categories without order (gender, colors )<br>
    b. Ordinal - Ordered categories (satisfaction levels: high, low, medium, neutral)<br><br>
2. Numerical /Quantitative<br>
    a. Discrete - countable numbers (number of children )<br>
    b. Continous - Measured values (weight , time )<br>

##### Characteristics 
1. The data contains only one variable 
2. The data focuses on describing only one variable therefore there is no relationship studied
3. Descriptive in nature i.e we use measures of central tendency on it alot 
4. It can be categorical and numerical 
5. Its represented using bar charts , histograms, piecharts and  boxplots .
6. It is the first step of analysis before bivariate and multivariate analysis 

##### Differences from other data types 
- Univariate data deals with one variable while bivariate data deals with two variables and Multivaraite deals with three or more varaibles

##### Methods for visualisation 
- Barchart  (shows categorical frequencies  ie insight on comparison of categories )
- Boxplot (shows median , quartiles  and outliers. Here we get insights on variablity and extreme value (outliers) )
- Histogram (shows frequency distribution of numerical data  and we get insight on skewness , spread and shape )
- Pie chart (proportion of categories ie part to whole relationships )

##### Measures of central tendency
- These measures tell us the centre of our data in simple terms 
<br>They include :

1. Mean - This is the arithmetic average . Its highly affected by outliers.
2. Median - This is the middle value when data is ordered 
3. Mode - This is the most frequent value in the data (This is useful with categorical data )

##### Measures of variability 
- These show us how much our data spreads around the center 
<br>Some of these measures include:

1. Range - Difference between the max and min value 
2. Variance - Average squared deviation from the mean 
3. Standard deviation (SD) - square root of the variance 
4. Interquartile range - This is the spread of the mean(50%) (Q3 -Q1)

#### MULTIVARIATE DATA RELATIONSHIPS 
- Multivariate data is data that involves 3 or more variables that are studied simultaneously 

##### Differences from other data types 
- It stuudies the complex dependencies between these variables while univariate data analysis is descriptive and bivariate data analysis is relationship based 

##### Dependence Techniques
- This is where statistical techniqs are used to sow how  one variable is dependent on other independent variables.
- This looks at how the independent variables cause change to the dependent varaible 

<h5>Common techniques here include: </h5>

1. Mulitple regression 
2. Logic regression 
3. MANOVA technique
4. Canonical Correlation Analysis
5. Structural Equation Modeling 

##### Interdependence techniques
- These are used when there is no clear disttinction between dependent and independent variables .
- We study the relationships, similarities and structures on all the variables simultaneously 

<h5>Main techniques include </h5>

1. Factor analysis - Used to identify factors that explain correlations among the variables 
2. Principle component Analysis - This method createds new uncorrelated variables from linear combinations of the original variables. This is used before regression 
3. Cluster Analysis - Groups variables into clusters such that objects with the same clusteer are more similar to each other thahn to those in other clusters .
4. Multidimensional Scaling(MDS) - represents data in low dimensional space for visualisation


##### Some of the applications of the common multivariate data analysis techniques 

1. Multiple Regression Analysis
- Its purpose is to predict thhe value of a dependent variables using two or more independent variables . Its main application is in marketing where it is used to predict sales based on advertising budget, price and competitor activity 
2. Logistic Regression 
- Its purpose is to estimate the probability of a categorical outcome from multiple independent variables . Its main use is in banking systems where it is used to predict a loan default based on income, credit score and debt of a given customer.
Other uses include predicting whether a patient has a disease or not based onthe test results 
3. Factor Analysis 
- Its purpose is to reduce many observed variables into fewer unobserved factors by identifying underlying dimensions . Its main use is in market research where it is used to identify key factors influencing consumer brand preference 
4. Cluster Analysis 
- This is mainly used in biology to classify species based on genetic similarity like Vertebrates and Invertebrates 
- Another use is in grouping of customers by purchasing behavior 
5. Multivariiate Analysis of Variance (MANOVA) 
- This extends ANOVA to compare groups means across multiple dependent variables simultaneously. Its main use is in medicine to evaluate the effect of a drug on blood pressure and heart rate at the same time .
- Another use is ineduccation to see how certain teaching methods affect scores in subjects like math
