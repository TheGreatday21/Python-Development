### **The random module** 

##### random choice

```python
### Using the random library 
import random # i can now access all the functions in this module 
from random import choice as ch #this syntax only gets a specific function and you can rename it to anything you want 

##starting off with the choice funtion 
coin = random.choice(['heads','tails']) #the function takes a list as a parameter with the choices in this case its a 50/50 choice 
coin2 = ch(['guns','blazing'])
print(coin)
print(coin2)
```

#### random int

```python
## This generates a random integer in a given range 
number = random.randint(1,10) ##generating integers from 1 included to 10

print(number)
```

#### random shuffle

```python
## this shuffles numbers randomly in the list it is given 
cards = ['jack','queen','king']

random.shuffle(cards)
#using a loop to print out these shuffles systematically 

for card in cards:
    print(card)#pretty cool to shuffle names in a long list like class group thingy lol
```

### **sys module**

```python
##This is the system module containing function to manupulate the system itself 

### argv - arguments directly in the terminal 

import sys 

print("Hello, my name is:", sys.argv[0])#the first argument in the list in the param of the argv function is the name of the file so thats why the index 1 will contain your name 
##index 0 is the name of the file 

print("Hello, my name is:", sys.argv[1])
#it will return what i type in . 

##the best ways to handle error of the user not placing in the name we code defensively using an exception or if conditional s coupled with the sys.exit function 

try :
    print("Hello, my name is:", sys.argv[2])
except IndexError:
    print("Too few arguments ")


if len(sys.argv) < 2:
    sys.exit("Too few arguments")
elif len(sys.argv) >2:
    sys.exit('Too many arguments')

print("Hello there: ",sys.argv[1])
```

### **API - Application Programming Interface**


```python
## These are third party services we write code to talk to to incorperate some data into our program 

import json ##this makes the json response prettier and easy to understand
import requests ##this is used for talking to urls and apis 
import sys ##this enables us to use command line arguments 


url = 'https://itunes,applle.com/search?entity=song&limit=1&term=weezer' #this is the itunes api to get 1 song for weezer
#we recieve a json file with the response to the serach we wanted - Its in json cause its quick 

url2 = 'https://itunes,applle.com/search?entity=song&limit=10&term='+ sys.argv[1] #this appends the term the user places in while initially running the program 

if len(sys.argv) != 2:
    sys.exit("Which musician's songs do you want?")

response = requests.get(url2) #this gets the json from the url for the musician the user wants

print(json.dumps(response.json(),indent=2))#printing out the response in json format . This almost looks like dictionary format 
#the dump function has the indent named parameter to specify how spaced the output will be 

########## -- we are going to change what response we get back to be more specific to what we want 

o = response.json() #storing the response from the server in a new variable called 0(object)

for result in 0["results"]: #there is a key in the dictionary in the response called results and that key only contains one value which is the song 
    print(result["trackName"])

```
